```python
def euclideanDistance(data_1, data_2):
    data_len=len(data_1)
#     test_len=len(data_2)
    dist = 0
   
    for i in range(data_len):
            dist = dist + np.square(data_1[i] - data_2[i])
    return np.sqrt(dist)
```

```python
def cityBlockDistance(data_1,data_2):
    data_len=len(data_1)
#     test_len=len(data_2)
    dist=0
    for i in range(data_len):
            dist =dist + np.abs(data_1[i]-data_2[i])
    return dist   
```

```python
def chessBoard(data_1,data_2):
    data_len=len(data_1)
    temp=[]
#     test_len=len(data_2)
    dist=0
    for i in range(data_len):
          temp.append(np.abs(data_1[i]-data_2[i]))
    return max(temp)  
```

```python
def cosineSimilarity(data_1, data_2):
    dot = np.dot(data_1, data_2)
#     print(dot)
    norm_data_1 = np.linalg.norm(data_1)
#     print(norm_data_1)
    norm_data_2 = np.linalg.norm(data_2)
#     print(norm_data_2)
    cos = dot / (norm_data_1 * norm_data_2)
    return (1-cos)
```

```python
def brayCurtis(data_1, data_2):
    data_len=len(data_1)
    numerator,denominator = 0, 0
    for i in range(data_len):
        numerator+=np.abs(data_1[i] -data_2[i])
        denominator += data_1[i]+data_2[i]
    return numerator/denominator
```

```python
def canberra(data_1,data_2):
    data_len=len(data_1)
    dist=0
    for i in range(data_len):
        num=np.abs(data_1[i] - data_2[i])
        den=np.abs(data_1[i])+np.abs(data_2[i])
        dist+=num/den
    return dist
```

```python
def correlation(data_1,data_2):
    data_len=len(data_1)
    d1_mean=np.mean(data_1)
    d2_mean=np.mean(data_2)
    den_1,den_2=0,0
    num_part=0
    for i in range(data_len):
        num_part+=(data_1[i]-d1_mean)*(data_2[i]-d2_mean)
        den_1+=np.square(data_1[i]-d1_mean)
        den_2+=np.square(data_2[i]-d2_mean)
    res=(num_part/np.sqrt(den_1*den_2))  
    return (1-res)
```

```python
def calcuDist(centroid,data_1):
    distance=[]
    classifier=[]
    distance_measures=["euclidean","city-block","chess-board","cosine","bray-curtis","canberra","correlation"]
    for d in distance_measures:
        dist=[]
        if d == "euclidean":
            
            for x in centroid:
                   dist.append(euclideanDistance(x,data_1))
            if dist.index(min(dist)) == 0 :
                 distance.append(min(dist))
                 classifier.append("Iris-setosa")
            elif dist.index(min(dist)) == 1:
                 distance.append(min(dist))
                 classifier.append("Iris-versicolor")
            else:
                 distance.append(min(dist))
                 classifier.append("Iris-virginica")
                    
        elif d == "city-block":
            for x in centroid:
                   dist.append(cityBlockDistance(x,data_1))
            if dist.index(min(dist)) == 0 :
                 distance.append(min(dist))
                 classifier.append("Iris-setosa")
            elif dist.index(min(dist)) == 1:
                 distance.append(min(dist))
                 classifier.append("Iris-versicolor")
            else:
                 distance.append(min(dist))
                 classifier.append("Iris-virginica")
                    
        elif d == "chess-board":
            for x in centroid:
                   dist.append(chessBoard(x,data_1))
#             print(dist)
            if dist.index(min(dist)) == 0 :
                distance.append(min(dist))
                classifier.append("Iris-setosa")
            elif dist.index(min(dist)) == 1:
                 distance.append(min(dist))
                 classifier.append("Iris-versicolor")
            else:
                 distance.append(min(dist))
                 classifier.append("Iris-virginica")
       
        elif d == "cosine":
            for x in centroid:
                   dist.append(cosineSimilarity(x,data_1))
#             print(dist)
#             print(dist.index(min(dist)))
            if dist.index(min(dist)) == 0 :
                distance.append(min(dist))
                classifier.append("Iris-setosa")
            elif dist.index(min(dist)) == 1:
                 distance.append(min(dist))
                 classifier.append("Iris-versicolor")
            else:
                 distance.append(min(dist))
                 classifier.append("Iris-virginica")
        
        elif d == "bray-curtis":
            for x in centroid:
                   dist.append(brayCurtis(x,data_1))
            if dist.index(min(dist)) == 0 :
                distance.append(min(dist))
                classifier.append("Iris-setosa")
            elif dist.index(min(dist)) == 1:
                 distance.append(min(dist))
                 classifier.append("Iris-versicolor")
            else:
                 distance.append(min(dist))
                 classifier.append("Iris-virginica")
        elif d =="canberra":
            for x in centroid:
                    dist.append(canberra(x,data_1))
            if dist.index(min(dist)) == 0 :
                distance.append(min(dist))
                classifier.append("Iris-setosa")
            elif dist.index(min(dist)) == 1:
                 distance.append(min(dist))
                 classifier.append("Iris-versicolor")
            else:
                 distance.append(min(dist))
                 classifier.append("Iris-virginica")
        
        elif d =="correlation":
            for x in centroid:
                    dist.append(correlation(x,data_1))
            if dist.index(min(dist)) == 0 :
                distance.append(min(dist))
                classifier.append("Iris-setosa")
            elif dist.index(min(dist)) == 1:
                 distance.append(min(dist))
                 classifier.append("Iris-versicolor")
            else:
                 distance.append(min(dist))
                 classifier.append("Iris-virginica")
    return classifier,distance
   
```
