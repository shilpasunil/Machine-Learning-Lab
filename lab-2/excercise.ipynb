```python
import numpy as np
import glob
import os
from PIL import Image, ImageChops 
import pandas as pd
import matplotlib.image as mpimg 
import matplotlib.pyplot as plt
```

```python
array=[1,2,3,8,9]
matrix=[[1,2,3,8,9],[4,5,6,9,8],[9,8,7,1,7],[2,11,3,4,5],[4,6,7,8,9]]
print("array is initialised to:")
print(array)
print("matrix is initialised to:")
matrix
```

```python
print("Mean of array")
print(np.mean(array))
print("Standard deviation of array")
print(np.std(array))
```

```python
print("Transpose of the matrix:")
print(np.transpose(matrix))
print("Inverse of matrix:")
print(np.linalg.inv(matrix))
```

```python
print("After adding array as the last row of the matrix:")
print(np.append(matrix,np.array([array]),axis=0))
```
