IMPORT NECCESARY PACKAGES

```python
import numpy as np
import matplotlib.pyplot as plt
```

INITIALIZE MEAN, COV1,COV2,COV3,COV4 AND COV5 (GIVEN)

```python
mean = [0,0]
cov1 = [[0.2, 0], [0, 2]]
cov2 =[[2,0],[0,0.2]]
cov3=[[1,0.5],[0.5,1]]
cov4=[[0.3,0.5],[0.5,2]]
cov5=[[0.3,-0.5],[-0.5,2]]
```

GENERATE 500 2 DIMENSIONAL GUASSIAN VECTOR FROM MEAN, COV1 AND PLOT THOSE VECTORS

```python

x, y = np.random.multivariate_normal(mean, cov1, 500).T
plt.plot(x,y,'x')
plt.title("FIRST CASE OF COVARIANCE MATRIX")
plt.axis('equal')
plt.xlabel('X-cordinate of gaussian vector')
plt.ylabel('Y-cordinate of gaussian vector')
plt.show()
```

GENERATE 500 2 DIMENSIONAL GUASSIAN VECTOR FROM MEAN, COV2 AND PLOT THOSE VECTORS

```python

x, y = np.random.multivariate_normal(mean, cov2, 500).T
plt.plot(x,y,'.',color="red")
plt.title("SECOND CASE OF COVARIANCE MATRIX")
plt.axis('equal')
plt.xlabel('X-cordinate of gaussian vector')
plt.ylabel('Y-cordinate of gaussian vector')
plt.show()
```

GENERATE 500 2 DIMENSIONAL GUASSIAN VECTOR FROM MEAN, COV3 AND PLOT THOSE VECTORS

```python
x, y = np.random.multivariate_normal(mean, cov3, 500).T
plt.plot(x, y,'+',color="blue")
plt.title("THIRD CASE OF COVARIANCE MATRIX")
plt.axis('equal')
plt.xlabel('X-cordinate of gaussian vector')
plt.ylabel('Y-cordinate of gaussian vector')
plt.show()
```

GENERATE 500 2 DIMENSIONAL GUASSIAN VECTOR FROM MEAN, COV4 AND PLOT THOSE VECTORS

```python
x, y = np.random.multivariate_normal(mean, cov4, 500).T
plt.plot(x, y,'x',color="green")
plt.title("FOURTH CASE OF COVARIANCE MATRIX")
plt.axis('equal')
plt.xlabel('X-cordinate of gaussian vector')
plt.ylabel('Y-cordinate of gaussian vector')
plt.show()
```

```python
x, y = np.random.multivariate_normal(mean, cov5, 500).T
plt.plot(x, y,'+')
plt.title("FIFTH CASE OF COVARIANCE MATRIX")
plt.axis('equal')
plt.xlabel('X-cordinate of gaussian vector')
plt.ylabel('Y-cordinate of gaussian vector')
plt.show()
```
